## Germany Macroeconomic Indicators and Market Trends(2020-2025)

```python
plt.figure(figsize=(14,8))

# Plot GDP
plt.plot(data['date'], data['GDP'], label='GDP Growth (%)')
plt.legend()
plt.title("GDP Growth (2020–2025)")
plt.xlabel("Date", fontsize = 12)
plt.xticks(rotation=90)

plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,8))
# CPI
plt.plot(data['date'], data['CPI'], label='Inflation (CPI)')
plt.legend()
plt.title("Inflation CPI")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,8))


# Unemployment
plt.plot(data['date'], data['Unemployment'], label='Unemployment (%)')

plt.legend()
plt.title("Unemployment Percentage")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,8))
# Bond Yields
plt.plot(data['date'], data['Bond_Yield'], label='10Y Bond Yield')

plt.legend()
plt.title("Bond Yields")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,8))

# EUR/USD
plt.plot(data['date'], data['EURUSD'], label='EUR/USD')

plt.legend()
plt.title("EUR/USD")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,8))
# ZEW Sentiment
plt.plot(data['date'], data['ZEW_Sentiment'], label='ZEW Sentiment', color = 'magenta')
plt.legend()
plt.title("ZEW Sentiment")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.ylabel("Value")
plt.show()
```
